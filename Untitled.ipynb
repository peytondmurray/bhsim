```python
import bh
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import cmocean.cm as ccm
import matplotlib.transforms as mt
import scipy.interpolate as si
import pymc3 as pm
import out
plt.style.use('dark_background')
```

```python
th0 = 80
theta_0 = th0 * np.pi / 180

m = 1
n = [0, 1]
alpha = np.linspace(0, 2 * np.pi, 1000)
r = np.arange(6, 40, 0.5)

df = bh.generate_image(alpha, r, theta_0, n, m, max_steps=3)
```

```python
cmap = ccm.gray
fig, ax = plt.subplots(figsize=(20, 20), subplot_kw={'projection': 'polar'}, facecolor=cmap(0))
ax.set_axis_off()
ax.set_theta_zero_location('S')

for n_value in n[::-1]:
    df_n = df.loc[df['n'] == n_value]
    ax.scatter(df_n['alpha'], df_n['b'], c=df_n['flux'], cmap=cmap, marker='o')
```

```python
cmap = cm.viridis
fig, ax = plt.subplots(figsize=(20, 20))#, facecolor=cmap(0))

for n_value in n[:-1]:
    df_n = df.loc[df['n'] == n_value]
    ax.scatter(df_n['alpha'], df_n['b'], c=df_n['flux'], cmap=cmap)
```

```python
df_n = df.loc[df['n'] == 0]
f = si.interp2d(df_n['alpha'], df_n['b'], df_n['flux'], kind='cubic')
```

```python
with pm.Model() as model:
    alpha = pm.Uniform('alpha', 0, 2*np.pi)
    b = pm.Uniform('b', 6, 40)
    
    def logp():
        return f(alpha, b)
    
    flux_pdf = pm.DensityDist('flux_pdf', logp=f, observed=(df_n['alpha'], df_n['b']))
    trace = pm.sample(500, return_inferencedata=True)
```
