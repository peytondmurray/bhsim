```python
import bh
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
import out
plt.style.use('dark_background')
```

```python
expr_opz = bh.expr_one_plus_z()
expr_opz
```

```python
opz = sy.lambdify(['alpha', 'b', 'theta_0', 'M', 'r'], expr_opz)
```

```python
alpha = np.linspace(0, 2*np.pi, 1000)
th0 = 60
theta_0 = th0*np.pi/180
m = 1

a_arrs = []
b_arrs = []
z_arrs = []

for r in np.arange(6, 40, 0.5):
    n = 0
    a_arrs.append(alpha)
    b = bh.impact_parameter(alpha, r, theta_0, n, m, max_steps=3)
    b_arrs.append(b)
    z_arrs.append(opz(alpha, b, theta_0, m, r) - 1)
```

```python
a_arr = np.concatenate(a_arrs)
b_arr = np.concatenate(b_arrs)
z_arr = np.concatenate(z_arrs)
```

```python
fig, ax = plt.subplots(figsize=(20, 20), subplot_kw={'projection': 'polar'})
ax.set_axis_off()
ax.set_theta_zero_location('S')

for z in [-0.15, -0.1, -0.05, 0, 0.05, 0.10, 0.15, 0.2, 0.25, 0.5, 0.75]:
    mask = np.isclose(z_arr, z, rtol=0, atol=0.001)
    ax.plot(a_arr[mask], b_arr[mask], '.w')

ax.scatter(a_arr, b_arr, c=z_arr)
```
